## Ejercicio 1
Dada la siguiente lista:
> ```ejer_1 = [1,2,3,4,5]```

Inviértela para que quede de la siguiente manera
> ```ejer_1 = [5,4,3,2,1]```

````python
ejer_1 = [1,2,3,4,5]
invertida = ejer_1[::-1] # con [::-1] inverto el órden de la lista en nueva variable
print(ejer_1) # la variable inicial no cambia
print(invertida) # la variable nueva es la que cambia
````

Otra forma

````python
ejer_1 = [1,2,3,4,5]
otra_invertida = list(reversed(ejer_1)) # con reversed invierte directamente y pido que lo liste
print(ejer_1) # la variable inicial no cambia
print(otra_invertida) # la variable nueva es la que cambia
````

Otra forma

````python
ejer_1 = [1,2,3,4,5]
ejer_1.sort(reverse = True) # aquí invierto directamente la lista
print(ejer_1)
````

## Ejercicio 2
Eleva todos los elementos de la lista al cuadrado
> ```ejer_2 = [1,2,3,4,5]```

````python
ejer_2 = [1,2,3,4,5]
cuadrado = [] # creo una nueva lista en blanco para que entren los resultados aquí
for i in ejer_2:
    resultado = i**2 # bucle que eleve cada elemento al cuadrado
    cuadrado.append(resultado) # que vaya incluyendo cada cálculo en la nueva lista
print(cuadrado)
````

## Ejercicio 3
Crea una lista nueva con todas las combinaciones de las siguientes dos listas:
> ```ejer_3_1 = ["Hola", "amigo"]```
>
> ```ejer_3_2 = ["Que", "tal"]```

Obten el siguiente output:

```['Hola Que', 'Hola tal', 'amigo Que', 'amigo tal']```

````python
# Declaro las dos listas
ejer_3_1 = ["Hola", "amigo"]
ejer_3_2 = ["Que", "tal"]

# creo una lista nueva y voy concatenando cada elemento de cada lista según su posición
lista_output = (ejer_3_1[0] + " " + ejer_3_2[0]),(ejer_3_1[0]+ " " + ejer_3_2[1]),(ejer_3_1[1] + " " + ejer_3_2[0]),(ejer_3_1[1] + " " + ejer_3_2[1])
print(list(lista_output)) # imprimo la lista de la variable
````

````python
# otra forma de hacerlo, con for
ejer_3_1 = ["Hola", "amigo"]
ejer_3_2 = ["Que", "tal"]

resultado = []
for objeto in ejer_3_1:
    for objeto2 in ejer_3_2:
        resultado.append(f"{objeto}{objeto2}")
print(resultado)
````

````python
# la forma que nos enseña David
import itertools

ejer_3_results = []
for combination in itertools.product(ejer_3_1,ejer_3_2):
    string = ""
    for value in combination:
        string = string + " " + value
    ejer_3_results.append(string [1:])
print(ejer_3_results)
````

## Ejercicio 5
Dada la siguiente lista, elimina todos los valores iguales a 3
> ```ejer_5 = [3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3]```

TIP: No intentes eliminar elementos sobre la lista que estás iterando. Haz una copia con `ejer_5.copy()`.

````python
ejer_5 = [3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3]
sin_3 = ejer_5.copy() # hago una copia de la variable con .copy()

for i in ejer_5:
    if i ==3: # le digo que busque en la lista el numero 
        sin_3.remove(i) # cuando lo encuentra, entonces lo elimina
print(sin_3)
````

## Ejercicio 11
Convierte la siguiente lista en un diccionario
> ```ejer_11 = [("x", 1), ("x", 2), ("x", 3), ("y", 1), ("y", 2), ("z", 1)]```

TIP: Puedes usar dos variables de control en un bucle `for`. Una por cada valor de las tuplas.

````python
ejer_11 = [("x", 1), ("x", 2), ("x", 3), ("y", 1), ("y", 2), ("z", 1)]
diccionario ={} # creo un diccionario en blanco

for key,value in ejer_11: # las dos variables de conrol son key y value
    diccionario [key] = value # le digo que itere key y value y que, por cada clave nueva le incluya el valor por orden
print(diccionario) # como hay varias claves iguales, entiendo que se ha quedado con el último valor que le hemos dado, ha sobreescrito los anteriores.

print("-"*50)
# si directamente genero un dicccinario de la variable ejer_11 con dict, me daría el mismo resultado
dict(ejer_11)
````

````python
# esta es la mejor solución, porque guardamos todos los datos y nos los perdemos
ejer_11 = [("x", 1), ("x", 2), ("x", 3), ("y", 1), ("y", 2), ("z", 1)]
diccionario1 = {}
for clave, valor in ejer_11:
    if clave in diccionario1.keys():
        diccionario1[clave].append(valor)
    else:
        diccionario1[clave] = [valor]
print(diccionario1)
````
