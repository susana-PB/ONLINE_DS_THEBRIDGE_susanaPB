## Ejercicio 1
Escribe una función que convierta números del 1 al 7 en nombres de los dias de la semana. La función constará de un único argumento numérico y una salida de tipo string

````python
'''
Función que convierte los número del 1 al 7 en nombres de los días de la semana.
Tiene un único argumento numérico y una salida tipo string.

Input = introducir un numero del 1 al 7

Output = me devuelve el día de la semana que equivale al número introducido
'''

def num_dias(numero):
    dia_semana = ["Lunes", "Martes", "Miercoles", "Jueves", "Viernes", "Sabado","Domingo"]
    for x,y in enumerate(dia_semana): # con enumerate saco el indice y el valor  de la lista
        if numero -1 == x: # entendiendo que comenzamos en 1, así nos alineamos con el 0 de python
            return y # si el numero -1 es igual al numero del indice, que me devuelva el valor de su posicion
   
````

````python
num_dias(2)
````

## Ejercicio 3
Escribe una función que compare dos números. La función tiene dos argumentos y hay tres salidas posibles: que sean iguales, que el primero sea  mayor que el segundo, o que el segundo sea mayor que el primero

````python
def comparar_numeros(x,y): # con dos argumentos a comparar
    if x == y: # Aquí valoramos si son iguales
        return "Los números son iguales" # pedimos el resultado en cuanto se cumpla la condición
    elif x > y: # Aquí valoremos si x es mayor que y
        return f"{x} es mayor que {y}"
    elif x < y: # Aquí valoremos si x es menor que y
        return f"{x} es menor que {y}"
````

## Ejercicio 4
Escribe una función que sea un contador de letras. En el primer argumento tienes que introducir un texto, y el segundo que sea una letra que contar. La función tiene que devolver un entero con el número de veces que aparece esa letra, tanto mayúscula, como minúscula

````python
'''
Función que cuenta la letra que hay en un texto.
Tiene como argumentos texto y letra. 

Input = introducir un texto

Output = numero int con el numero de veces que aparece esa letra 
'''

def contador(texto,letra): # Queremos contar numero de letras de un texto, ponemos dos argumentos
    letra_buscada = letra.lower() # Definimos una variable que convierta el argumento letra en minuscula
    contador = 0 # Para que pueda contar las letras, tenemos que definir una variable a 0
    for caracter in texto.lower():
        if caracter == letra_buscada: # si coincide que la letra está en el texto:
            contador = contador +1 # Va sumando al contador por cada letra == que encuentre
    return contador
````

## Ejercicio 5
Escribe una función que tenga un único argumento, un string. La salida de la función tiene que ser un diccionario con el conteo de todas las letras de ese string, siendo la clave la letra y el valor el conteo.

````python
'''
Función que solo tiene un string.
Tiene como argumentos texto y letra. 

Input = introducir un string

Output = diccionario con el conteo de todas las letras del string (clave = letra y valor = conteo)
'''

def contar_letras(string):
    contando = {} # defino un diccionario que es donde entrará el conteo de las letras
    for caracter in string.lower(): # que busque cada letra en el string introducido, convertido en minuscula
        if caracter in contando: # si está el caracter
            contando[caracter] += 1 # sumamos al valor 1
        else:
            contando[caracter] = 1 # si no está, el valor es 1
    return contando
````

````python
contar_letras("Elefante")
````

## Ejercicio 6
Escribir una función que añada o elimine elementos en una lista. La función necesita los siguientes argumentos:
* lista: la lista donde se añadirán o eliminarán los elementos
* comando: "add" o "remove"
* elemento: Por defecto es None.

Tendrá como salida la lista

````python
'''
Función que añada o elimine elementos de una lista.

Input = lista, comando add/remove y elemento (None)

Output = lista
'''

def añadir_eliminar(lista,comando,elemento):
    comando = comando.lower() # genero el comando para que convierta en minuscula
    if comando == "add":
        lista.append(elemento) # que añada el elemento a la lista
    elif comando == "remove":
        lista.remove(elemento) # que elimine el elemento de la lista
    else:
        return f"El elemento '{elemento}' no está en la lista" # que avise si no está en la lista
    return lista
````

````python
añadir_eliminar(["elemento1", "elemento2"], "add", "Pez")
````

## Ejercicio 9
Define en una única celda las siguientes funciones:
* Función que calcule el área de un cuadrado
* Función que calcule el área de un triángulo
* Función que calcule el área de un círculo

En otra celda, calcular el area de:
* Dos círculos de radio 10 + un triángulo de base 3 y altura 7
* Un cuadrado de lado = 10 + 3 círculos (uno de radio = 4 y los otros dos de radio = 6) + 5 triángulos de base = 2 + altura = 4

````python
'''
Funcion que calcula el área de un cuadrado -> Área = lado x lado
Función que calcula el área de un triángulo -> Área = base x altura
Función que calcula el área de un círculo -> Área = pi x r^{2}

Input = lnumero de palabras

Output = frase completa, separando las palabras con espacios.
'''

import math
def area_cuadrado(lado):
    area_cuadrado = float(lado) * float(lado)
    return area_cuadrado
def area_triangulo(base,altura):
    area_triangulo = float(base) * float(altura)
    return area_triangulo
def area_circulo(radio):
    area_circulo = math.pi * float(radio**2)
    return area_circulo
````

````python
print(area_cuadrado(5))
print(area_triangulo(5,10))
print(area_circulo(15))
````
