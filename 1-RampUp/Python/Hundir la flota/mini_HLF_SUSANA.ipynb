## Mini Hundir la Flota
#### Primeros pasos
Vamos a empezar a programar trozos de código que nos servirán para el HLF:

* Crear un Tablero como una lista de listas de 10x10 que contenga caracteres espacio: " "
* Insertar dos barcos sustituyendo en el Tablero los espacios en blanco por el caracter "B":
    * Uno de 4 casillas en horizontal: casillas [1][0], [1][1], [1][2], [1][3]
    * Uno de 3 casillas en vertical:   casillas [3][3], [4][3], [5][3]
* Escribir una celda de código separada de las anteriores para simular que el usuario está efectuando un disparo: 
    *  Solicite al usuario dos coordenadas `i,j` mediante dos inputs
    *  Compruebe si esas coordenadas están ocupada por una casilla barco (que haya una en "B")
        * Si hay una "B":
            * Cambiar el contenido de esa posición por una "x"
            * Imprimir por pantalla el texto "Tocado en posición `i,j`"
        * Si hay un " ":
            * Cambiar el contenido de esa posición por una "o"
            * Imprimir por pantalla el texto "Agua"
    * Imprimir por pantalla el Tablero
* Para comprobar que funciona bien la celda de disparar ejecutarla varias veces metiendo distintas coordenadas

### Bonus
* En la celda de disparar, si el usuario acierta a tocar un barco, vuelve a disparar
* Resuélvelo con funciones:
    * Una para solicitar una coordenada válida a la que llamarás dos veces, una par i y otra par j
    * Una `disparar` que devuelva `True` si el usuario ha acertado o `False` si no.
    


```python
'''
El objetivo de este ejercicio es simular el juego de Hundir la flota de una manera sensilla.

El input es:
Se introducen las coordenadas i,j

El output:
Si las coordenadas coiniden con la casilla que contiene la B: 
1 - En esa casilla la B pasará a ser una X
2 - Se imprimirá en pantalla: Tocado en posición i,j
Si las coordenadas contienen un espacio " ":
1 - Esa casilla pasará a ser una o
2 - Se imprimirá en pantalla: Agua
El tablero se debe imprimir en pantalla

Reglas/Restricciones:
- Las coordenadas son numeros int positivos
- Las coordenadas debe escribirlas el usuario
- Si la coordenada introducida contiene la B, vuelve a disparar

'''
```

```python
'''
Mi Pseudocodigo:
1 - Creo un tablero de 10x10
2 - Inserto los barcos en las coordenadas indicadas
3 - Con una función comprueba si las coordenadas están en casilla con B o no y ejecuta en cada caso
4 - Con una función genero un bucle que permite disparar si se acierta en la X
5 - COMIENZA EL JUEGO!
'''
```

```python
# 1 - Creo un tablero de 10x10

tablero = [] # genero tablero en blanco
filas = 10 
columnas = 10

for fila in range(filas):
    nueva_fila = []
    for columna in range(columnas):
        nueva_fila.append(" ")
    tablero.append(nueva_fila)

print("\nTablero") # Imprimir el tablero
for fila in tablero:
    print(" | ".join(fila))
```

```python
# 2 - Inserto los barcos en las coordenadas indicadas

def colocar_barcos(tablero):
    tablero[1][0] = "B"
    tablero[1][1] = "B"
    tablero[1][2] = "B"
    tablero[1][3] = "B"
    tablero[3][3] = "B"
    tablero[4][3] = "B"
    tablero[5][3] = "B"
    return tablero
print(colocar_barcos(tablero))
print("\nTablero")  
for fila in tablero:
    print(" | ".join(fila))
```

```python
# 3 - Con una función comprueba si las coordenadas están en casilla con B o no y ejecuta en cada caso
def adivinar_celda(tablero,i,j):
    i_real = i-1 # para coordinar la numeración con Python 
    j_real = j-1
    try:
        celda = tablero[i_real][j_real] # genero variable con las coordenadas
    except IndexError:
        return "Coordenadas fuera de los límites (1-10). ¡Disparo inválido!"   
    
    if celda == "B":
        tablero[i_real][j_real] = "X"
        return "Tocado"    
    elif celda == " ":
        tablero[i_real][j_real] = "o"
        return "Agua"
    else:
        return f"Ya habías disparado en ({i},{j})."

# 4 - Con una función genero un bucle que permite disparar si se acierta en la X
def jugar(tablero):
    turno_repetido = True
    while turno_repetido:
        try:
            i = int(input("Dame una coordenada de fila (1-10): "))
            j = int(input("Ahora dame una coordenada de columna (1-10): "))
            resultado = adivinar_celda(tablero,i,j)
            print(resultado)
            
            if "Tocado" in resultado:
                continue
            else:
                turno_repetido = False
        except ValueError:
            print("Entrada inválida. Intenta de nuevo.")
            break
```

```python
# 5 - COMIENZA EL JUEGO!

jugar(tablero)
print()

print("\nTablero Actualizado")  
for fila in tablero:
    print(" | ".join(fila))
```
