## Juego del Ahorcado
El funcionamiento es que vaya mostrando por pantalla ```_ _ _ _ _``` las rayas bajas correspondientes al número de letras que tiene la palabra a adivinar. El usuario deberá insertar una letra y, si esa letra existe, se mostrará o mostrarán las letras acertadas ```S _ S _ _```. 

Si falla, se mostrará el número de errores que lleva y, si lleva 6 errores, entonces se mostrará "Fin del juego".
Encapsúlalo en una función.

````python
'''
OBJETIVO: adivinar una palabra diciendo sus letras. 
--> Si la letra está en la palabra, aparecerá reflejada, si no está en la palabra, pierde una oportunidad. 
--> Hay 6 oportunidades para acertar.

*Input: 
--> introducir una palabra.
--> introducir letras para adivinar la palabra.

*Output: 
--> Mostrará el número de errores en caso de fallo. 
--> Mostrará la palabra en su posición en caso de haber acertado. 
--> Mostrará "Fin de juego"

'''
# Introducción de la palabra con input
palabra_secreta = list(input("Dime una palabra.").upper()) # pido input para conseguir la variable en MAYUSCULA

def ahorcado(palabra_secreta): # hago una función que cuenta las letas de la palabra
    huecos_palabra = ["_"] * len(palabra_secreta) # genero una lista que ponga tantos _ como letras tiene la palabra 
    return huecos_palabra # pongo return para que me devuelva los huecos de la palabra

resultado_actual = ahorcado(palabra_secreta) 
print(palabra_secreta)
print(" ".join(resultado_actual)) # aquí le pido que ponga un espacio vacío por cada espacio
````

````python
num_intentos = 6 # defino intentos

def adivinando(letra_intento,palabra_secreta,resultado_actual): # la función trabaja en 3 valores
    fallo = False
    if letra_intento in palabra_secreta:
        for i in range(len(palabra_secreta)):
            if palabra_secreta[i] == letra_intento:
                resultado_actual[i] = letra_intento
    else:
        fallo = True
        print(f"La letra '{letra_intento}' no se encuentra en la palabra")
    return resultado_actual,fallo
````

````python
while num_intentos > 0 and "_" in resultado_actual:
    letra_intento = (input("Dame una letra: ")).upper() # con input pido una letra y la defino como variable       
    resultado_actual,fallo = adivinando(letra_intento,palabra_secreta, resultado_actual)
    if fallo == True:
        num_intentos = num_intentos - 1
        print(f"Intento fallido, te quedan {num_intentos} intentos.")
    else:
        print(f"Has acertado -> {resultado_actual}")
        print(" ".join(resultado_actual))

if num_intentos == 0: 
    print ("Fin de juego")
else:
    print("Has ganado!")
````
