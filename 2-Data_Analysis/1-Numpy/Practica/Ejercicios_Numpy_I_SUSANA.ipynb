## Ejercicio 4
Extrae todos los impares de `my_array`

````python
impares = my_array[my_array % 2 != 0]
print(impares)
````

## Ejercicio 5
En un nuevo array, sustituye todos los impares de `my_array` por -1

````python
nuevo_array = impares[impares % 2 != 0] = -1
print(nuevo_array)
````

## Ejercicio 6
En un nuevo array, sustituye todos los impares de `my_array` por -1. Esta vez usa `where`

````python
nuevo_array_2 = np.where(nuevo_array % 2 != 0, -1,nuevo_array)
print(nuevo_array_2)
````

## Ejercicio 8
Concatena los siguientes arrays verticalmente

```Python
a = np.arange(10).reshape(2,-1)
b = np.repeat(1, 10).reshape(2,-1)
```

````python
a = np.arange(10).reshape(2,-1)
b = np.repeat(1, 10).reshape(2,-1)

concatenada_v = np.vstack((a, b))
print(concatenada_v)
````

## Ejercicio 9
Concatena los siguientes arrays horizontamente

```Python
a = np.arange(10).reshape(2,-1)
b = np.repeat(1, 10).reshape(2,-1)
```

````python
concatenada_h = np.hstack((a, b))
print(concatenada_h)
````

## Ejercicio 10
Encuentra todos los elementos en común entre ambos arrays. [Para ello usa la función `intersect1d`](https://numpy.org/doc/stable/reference/generated/numpy.intersect1d.html)

```Python
a = np.array([1,2,3,2,3,4,3,4,5,6])
b = np.array([7,2,10,2,7,4,9,4,9,8])
```

````python
a = np.array([1,2,3,2,3,4,3,4,5,6])
b = np.array([7,2,10,2,7,4,9,4,9,8])

elem_comun = np.intersect1d(a, b)
print(elem_comun)
````
