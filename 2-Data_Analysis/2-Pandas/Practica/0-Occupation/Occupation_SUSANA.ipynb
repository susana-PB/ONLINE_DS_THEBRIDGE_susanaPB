### 18. ¿Cuál es la edad menos frecuente?

```python
# Primero cuento los valores age
conteo = users["age"].value_counts()
print(conteo)

# Después busco la edad menos fecuente
menos_frecuente = conteo[conteo == conteo.min()].index.tolist()
print(menos_frecuente)
```
