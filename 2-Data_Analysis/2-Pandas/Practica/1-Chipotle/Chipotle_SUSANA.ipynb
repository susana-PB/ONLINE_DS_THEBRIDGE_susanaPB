### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv) and assign it to a variable called chipo.

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep="\t")
chipo
```

### Step 3. Name of the max valued product

```python
chipo.info()
```

```python
# Dejar la columna item_price a int.
# Reemplazar $ por nada

chipo['item_price'] = chipo['item_price'].str.replace('$', '', regex=False).astype(float)
print(chipo['item_price'])

# # Mostrar el valor maximo
print(chipo["item_price"].max())
```

### Step 4. How many products cost more than $10.00?

```python
# Primero calculo el precio unitario y lo añado a la columna
precio_unitario = chipo['precio_unitario'] = chipo['item_price'] / chipo['quantity']
print(chipo[['item_name', 'quantity', 'item_price', 'precio_unitario']])
```

```python
# ahora busco el precio minimo unitario de cada producto, entenderé que es la base
# reset_index() que guarda item_name como una columna normal 
precio_minimo = chipo.groupby('item_name')['precio_unitario'].min().reset_index() 
print(precio_minimo)
```

```python
# Los productos que cuestan mas de $10.00
productos_mas_10 = precio_minimo[precio_minimo["precio_unitario"] >= 10.00]
print(productos_mas_10)
```

### Step 4.1: Y cuántos pedidos se han hecho con un producto de más de 10$? Es lo mismo?

```python
# Para calcularlo tomo en cuenta la variable precio_minimo
numero_pedidos = productos_mas_10.groupby('item_name').size().reset_index(name='precio_unitario')
print(numero_pedidos)

# NO ES LO MISMO:
    # Productos cuyo coste es superior a $10 -> hablamos de PRODUCTOS y calculamos sobre el producto MINIMO. Da como resultado PRODUCTOS
    # Pedidos realizados con producto de coste superior a $10 -> hablamos de NUMERO DE PEDIDOS y calculamos pedidos de productos de mas de %10. Da como resultado PEDIDOS.
```

### Step 4.2: Y cuántos pedidos se han hecho de más de 10$? Es lo mismo?

```python
# Calculamos los PEDIDOS que tienen precio unitario > 10
# NO ES LO MISMO:
    # Aquí hablamos de los PEDIDOS el precio de producto (item_price), no el precio unitario. Entiendo que una fila es 1 pedido.
pedidos_mas_10 = chipo[chipo['item_price'] > 10.0]
print(pedidos_mas_10)
```

### Step 4.3: Y en cuántos pedidos se ha pagado más de 10$ por un mismo producto? Es lo mismo?

```python
# Calculamos el número de PEDIDOS en los que se ha pagado mas de $10 por un mismo producto
mas_10_mismo_prod = chipo[chipo['precio_unitario'] > 10].shape[0]
print(mas_10_mismo_prod)
# NO ES LO MISMO:
    # Aquí calculamos el número total de pedidos (filas del DataFrame) que cumplen con esa condición.
```

### Step 5. What is the price of each item and name it unit_price. Get only item_name and unit_price

```python
# Ya tenía el precio unitario calculado más arriba, por lo que reemplazo uno por otro
chipo.rename(columns={'precio_unitario': 'unit_price'}, inplace=True)

print(chipo[['item_name', 'unit_price']])
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values('item_name')
```

### Step 7. What was the quantity of the most expensive item ordered? 2 ways

```python
# Calculamos la cantidad con max() que nos devuelve la cantidad mas alta

print(chipo['quantity'].max())
```

```python
# Con nlargest saco la fila que tiene mayor cantidad. Devuelve la fila completa.
print(chipo.nlargest(1, 'quantity'))
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
# Busco el item que sea == Veggie Salad Bowl y le digo que cuent los pedido (order_id)
veggie_salad = chipo[chipo['item_name'] == 'Veggie Salad Bowl']['order_id'].count()
print(veggie_salad)
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
# Primero confirmo que Canned Soda existe en la DataFrame
existe = ('Canned Soda' in chipo['item_name'].values)
print(existe)

# Luego busco las filas donde aparece Canned Soda y la cantidad pedida sea > 1, y cuento los pedidos.

canned_soda = chipo[(chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)]['order_id'].count()
print(canned_soda)
```
