## Ejercicio 12
¿Existe alguna relación lineal entre las variables numéricas del dataset? Intenta pintar este razonamiento mediante un gráfico

## Ejercicio 13
Representa el nivel de alcohol (columna `abv`) en un histograma. ¿Se trata de una distribución simétrica o asimétrica?

```python
beer['abv'].describe()
```

```python
beer['abv'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title("Histograma de IBU")
plt.xlabel("IBU")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Pequeña asimetría positiva (cola a la derecha = valores altos). La media es ligeramente mayor que la mediana.
# 50% es 0.056000 lo que dice que la mitad central está bastante concentrada alrededor de la mediana.
# std -> 0.013542 indica baja dispersión relativa respecto a la media, la mayoría de valores está cerca de 0.056–0.067.
```
