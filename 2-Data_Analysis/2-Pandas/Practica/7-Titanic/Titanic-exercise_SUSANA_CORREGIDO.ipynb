9. What is the name of the oldest person who died in the Titanic? Was he or she travelling alone or had any family travelling with them?  



In order to give an answer to the second question you should find out which columns give you that info. Usually part of your job as a Data Scientist will be get to know the dataset which you are working with. In this case the columns which give you that info are the following: 
        - 'sibsp' Number of Siblings/Spouses Aboard
        - 'parch' Number of Parents/Children Aboard

```python
# ESTE CODIGO SOLAMENTE ME DA EL VALOR DE LA PERSONA MAS MAYOR QUE VIAJABA SOLA

# Name of the oldest person
oldest_person = titanic.loc[titanic['age'].idxmax(), 'name']
print('El nombre de la persona más mayor es:', oldest_person)

# Viajaba solo? o tenia algún falimiar viajendo con él?
if oldest_person in titanic['name'].values:
    family = titanic.loc[titanic['name'] == oldest_person, ['sibsp', 'parch']].values[0]
    if family[0] == 0 and family[1] == 0:
        print(f'{oldest_person} viajaba solo.')
    else:
        print(f'{oldest_person} no viajaba solo, tenía familiares a bordo.')
```

```python
# ESTE CODIGO ES EL CORRECTO, ME DICE LA PERSONA MAS MAYOR QUE MURIO

# Name of the oldest person
oldest_person = titanic[titanic['survived'] == 0].loc[titanic[titanic['survived'] == 0]['age'].idxmax(),'name']
print('El nombre de la persona más mayor es:', oldest_person)
```

```python
# Viajaba solo? o tenia algún falimiar viajendo con él?
if oldest_person in titanic['name'].values:
    family = titanic.loc[titanic['name'] == oldest_person, ['sibsp', 'parch']].values[0]
    if family[0] == 0 and family[1] == 0:
        print(f'{oldest_person} viajaba solo.')
    else:
        print(f'{oldest_person} no viajaba solo, tenía familiares a bordo.')
```

Don't panic and read the documentation!




```python
counted = titanic[titanic['survived'] == 1].pivot_table(values= 'survived',index=['pclass','gender'],aggfunc='count')
counted
```

25. Add a new column with the sum of survived men and women, call it `counted['total']` 


```python
counted = titanic[titanic['survived'] == 1].pivot_table(index='pclass',columns='gender',values= 'age',aggfunc='count')
counted['total'] = counted['female'] + counted['male']
counted
```

34. Using [`kdeplot`]("https://seaborn.pydata.org/generated/seaborn.kdeplot.html") from `seaborn` represent those who not survived distributed by age.<>

Hint: First you should "filter" the `titanic` dataset where the column "survived" is 0, indexing the column `"age"` only.

Arguments you should pass to the function: 

    - color = "red"
    - label = "Not Survived"
    - shade = True
    
You should get something like this: 

![kdeplot](./img/kdeplotsur.png)

```python
ages_not_survived = titanic.loc[titanic['survived'] == 0, 'age'].dropna()
sns.kdeplot(
    data=ages_not_survived,
    color="red",
    label="Not Survived",
    shade=True)
plt.xlabel("Age")
plt.ylabel("Density")
plt.title("Distribución de edades de los pasajeros que no sobrevivieron")
plt.legend()
plt.show()
```
