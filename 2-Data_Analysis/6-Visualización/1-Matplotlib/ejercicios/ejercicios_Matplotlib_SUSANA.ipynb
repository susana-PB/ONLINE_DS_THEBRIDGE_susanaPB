2. Sabemos que podemos pintar gráficas de dos formas: la figura incluye los ejes o teniendo figura y ejes por separado.

Usando solo una figura, usa numpy para los valores del eje X entre 0 y 5. 

Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). 

Elige la precisión en el eje X que desees.  

```python
# Valores de X entre 0 y 5 con la precisión deseada
x = np.linspace(0, 5, 100)

# Recta con pendiente positiva de 3 que pasa por (0,0)
y1 = 3 * x

# Recta con pendiente negativa de 3 que pasa por (0, -5)
y2 = -3 * x - 5

# Crear figura con dos subgráficas (una fila, dos columnas)
plt.figure(figsize=(10, 4))
```

```python
# Subgráfica izquierda: recta con pendiente positiva de 3 que pase por (0,0)
plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title("Recta positiva")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)

# Subgráfica derecha: recta con pendiente negativa de 3 que pase por (0,-5)
plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title("Recta negativa")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)

plt.tight_layout()
plt.show()
```

5. Usando una figura Y EJES POR SEPARADO, usa numpy para los valores del eje X entre 0 y 5. Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). Elige la precisión en el eje X que desees.  

```python
# Valores de X entre 0 y 5
x = np.linspace(0, 5, 100)

# Recta con pendiente positiva de 3 que pasa por (0,0)
y1 = 3 * x

# Recta con pendiente negativa de 3 que pasa por (0, -5)
y2 = -3 * x - 5

# Crear figura
fig = plt.figure(figsize=(10, 4))

# Crear ejes por separado (izquierda y derecha)
ax1 = fig.add_axes([0.07, 0.15, 0.4, 0.75])   # [left, bottom, width, height]
ax2 = fig.add_axes([0.55, 0.15, 0.4, 0.75])

# Gráfica izquierda
ax1.plot(x, y1)
ax1.set_title("Recta positiva")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.grid(True)

# Gráfica derecha
ax2.plot(x, y2)
ax2.set_title("Recta negativa")
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.grid(True)

plt.show()
```

14. Ahora pintamos un gráfico de línea con las ventas respecto al año. Pon un título y etiquetas en los ejes.
Cambia el estilo a 'seaborn-white'. La línea debe ser con rayas y verde.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos
data = {
    "Year": [2015, 2016, 2017, 2018, 2019],
    "Sold_items_A": [1000, 3500, 4000, 5500, 7000]
}
df = pd.DataFrame(data)

# Aplicar estilo Seaborn general
plt.style.use('default')  # estilo seguro y disponible

# Gráfico de línea
plt.plot(df["Year"], df["Sold_items_A"], color="green", linestyle="--", marker='o')

# Etiquetas y título
plt.xlabel("Año")
plt.ylabel("Artículos vendidos")
plt.title("Ventas de artículos A por año")

plt.grid(True)
plt.show()
```

15. Otro departamento B ha vendido en esos años 2000, 3100, 5000, 4000, 6000 unidades. Incluye esa columna en el DataFrame y pinta en la misma gráfica las dos líneas. B es una línea punteada y roja. Muestra una leyenda abajo a la derecha.

```python
# incluir columna en dataframe con ventas de artículos B
df["Sold_items_B"] = [2000,3100,5000,4000,6000]

# Estilo por defecto
plt.style.use('default')

# Gráfico de línea para A
plt.plot(df["Year"], df["Sold_items_A"], color="green", linestyle="--", marker='o', label="Departamento A")

# Gráfico de línea para B
plt.plot(df["Year"], df["Sold_items_B"], color="red", linestyle=":", marker='s', label="Departamento B")

# Etiquetas y título
plt.xlabel("Año")
plt.ylabel("Artículos vendidos")
plt.title("Ventas de los Departamentos A y B por año")

# Grid
plt.grid(True)

# Leyenda abajo a la derecha
plt.legend(loc="lower right")

plt.show()
```

19. Vamos a pintar un histograma de unas alturas generadas aleatoriamente. Crea 300 alturas entre 150 y 190 cm.

Pista: https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.DataFrame.plot.html

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar 300 alturas aleatorias entre 150 y 190 cm
alturas = np.random.uniform(150, 190, 300)

# Crear histograma
plt.hist(alturas, bins=15, color='skyblue', edgecolor='black')  # 15 bins para distribuir las alturas

plt.xlabel("Altura (cm)")
plt.ylabel("Frecuencia")
plt.title("Histograma de alturas aleatorias")

plt.grid(axis="y")
plt.show()
```
